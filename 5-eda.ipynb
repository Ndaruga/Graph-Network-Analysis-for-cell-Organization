```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

features_df = pd.read_csv('extracted_features.csv', index_col='Unnamed: 0')
features_df.tail(15)
```

```python
features_df = features_df.loc[(features_df.iloc[:, :-1] != 0).any(axis=1)]
features_df.tail(15)
```

```python
features_df.shape
```

```python

# Remove the 'label' column for visualization purposes
if 'label' in features_df.columns:
    features_without_label = features_df.drop(columns=['label'])
else:
    features_without_label = features_df
```

```python
# 1. Feature Distributions - Histograms
plt.figure(figsize=(12, 8))
for feature in features_without_label.columns:
    sns.histplot(features_without_label[feature], kde=True, label=feature, alpha=0.6)
plt.xlabel('Feature Value')
plt.ylabel('Count')
plt.title('Feature Distributions')
plt.legend()
plt.show()
```

```python
# 2. Feature Correlations - Heatmap
correlation_matrix = features_without_label.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# 3. Class Distribution (if 'label' column exists) - Bar Chart
if 'label' in features_df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='label', data=features_df)
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Class Distribution')
    plt.show()
```

```python
# You can also explore scatter plots, box plots, or dimensionality reduction techniques for more in-depth analysis.
# For example, to visualize pairwise relationships between features using scatter plots:
sns.pairplot(features_df, hue='label' if 'label' in features_df.columns else None)
plt.show()
```

```python
# Or for box plots of features grouped by class (if 'label' column exists):
if 'label' in features_df.columns:
    plt.figure(figsize=(12, 8))
    for feature in features_without_label.columns:
        sns.boxplot(x='label', y=feature, data=features_df)
    plt.xlabel('Class')
    plt.ylabel('Feature Value')
    plt.title('Feature Distributions by Class')
    plt.show()
```
